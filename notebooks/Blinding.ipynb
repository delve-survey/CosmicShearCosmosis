```python
import astropy.io.fits as pf
import pylab as mplot
%pylab inline
import numpy as np
```

```python
dv_dir = '/project/chihway/chihway/CosmicShearCosmosis/datavectors/'
original = pf.open(dv_dir+'baseline_Jan12th24.fits')
blinded = pf.open(dv_dir+'baseline_Jan12th24_BLINDED.fits')
```

```python
original.info()
```

```python
mplot.figure(figsize=(10,4))
mplot.plot(original['xip'].data['value'], label='original')
mplot.plot(blinded['xip'].data['value'], label='blinded')
mplot.yscale('log')
mplot.ylabel('xip')
```

```python
mplot.figure(figsize=(10,4))
mplot.plot(original['xip'].data['value'], label='original')
mplot.plot(blinded['xip'].data['value'], label='blinded')
mplot.yscale('log')
mplot.ylabel('xip')
mplot.legend()
```
